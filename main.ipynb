```python
import requests
import pandas as pd
from plotly import graph_objs as go
from plotly.subplots import make_subplots
import seaborn as sns
import matplotlib.pyplot as plt
from config import client_id
import json

symbol = 'TSLA'
start_date = ''
end_date = ''
columns = ['open', 'high', 'low', 'close', 'volume', 'datetime']

# define an endpoint with a stock of your choice, MUST BE UPPER
endpoint = r"https://api.tdameritrade.com/v1/marketdata/{}/pricehistory".format(symbol)

# Example: For a 2 day / 1 min chart, the values would be:
# period: 2
# periodType: day
# frequency: 1
# frequencyType: min

# define the payload
payload = {'apikey' : client_id,
            'period' : '1',
            'periodType' : 'year', # day, month year
            'frequency' : '1',
            'frequencyType' : 'daily', 
           #'endDate':'',
           #'startDate':'',
            'needExtendedHoursData' : 'False'}

# make a request
content = requests.get(url = endpoint, params = payload)

# convert it dictionary object
data = content.json()
#data
```
